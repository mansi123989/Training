**Encoding**

* encoding refers to changing categorical data into numerical data
* types of encoding :- 1. ordinal encoding 2. label encoding 3. onehot encoding

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('laptop_data.csv')
data
```

```python
x = data[['Company']]
y = data[['TypeName']]
x_train , x_test, y_train, y_test = train_test_split(x , y ,test_size = 0.33 , random_state=2)
```

* 1. Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
en = OrdinalEncoder()
encode = en.fit_transform(x_train)
pd.DataFrame(encode, columns=['Company'])
```

```python
from sklearn.preprocessing import OrdinalEncoder
en = OrdinalEncoder()
encode = en.fit_transform(x_train)
```

* 2. Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
en = LabelEncoder()
encode = en.fit_transform(x_train)
pd.DataFrame(encode, columns=['TypeName'])
```

* 3. Onehot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
en1 = OneHotEncoder()
encode = en1.fit_transform(x_train)
pd.DataFrame(encode, columns=['TypeName'])
```
