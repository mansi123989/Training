**3D plot**

```python
from sklearn.datasets import make_regression
import plotly.express as px
import plotly.graph_objects as go
```

**genrate random dataset**

```python
x,y = make_regression(n_samples = 200 , n_features = 2 , n_informative = 2 , n_targets=1, noise = 50)
```

```python
import pandas as pd
df = pd.DataFrame({'feature1':x[: , 0] , 'feature2': x[: ,1] , 'target':y})
```

```python
fig = px.scatter_3d(df, x = 'feature1' , y ='feature2' , z = 'target')
```

```python
import pandas as pd
```

```python
data = pd.read_csv("laptop_data.csv")
```

**create new feature for operating system type**

```python
def os( operating_sys):
    if 'Windows' in operating_sys:
        return 'window'
    elif 'macOS' == operating_sys:
        return 'macOS'
    else:
        return 'other/linux'
    
```

```python
data['oper_sys'] = data['OpSys'].apply(os)
```
