```python
 **Standardization and Normalization**
```

```python
Standardization and Normalization are two common techniques used in data preprocessing, especially in machine learning, to rescale or transform features so they can be compared more easily or fed into algorithms that are sensitive to scale.
```

```python
**Standardization (Z-score normalization)**
```

```python
* Definition: Transforms the data to have a mean of 0 and a standard deviation of 1.
* Formula: z = x−μσ
* x: original value
* μ: mean of the feature
* σ: standard deviation of the feature
```

```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load your data
data = pd.read_csv('movies.csv')

# Select features
x = data[['budget']]  
y = data[['popularity']]

# Split the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=2)

# Standardize the feature
sc = StandardScaler()
scaled = sc.fit_transform(x_train)

# Convert to DataFrame
result = pd.DataFrame(scaled, columns=['budget'])
print(result.head())
```
