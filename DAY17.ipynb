```python
import numpy as np
```

```python
def step (z):
    return 1 if z>0 else 0
```

```python
def perceptron(x,y):
    x = np.insert(x,0,1,axis = 1)
    weights = np.ones(x.shape[1])
    ir = 0.1

    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(x[j] , weights))
        weights = weights +ir*(y[j] - y_hat) * x[j]
    return weights[0],weights[1:]    
```

```python
intercept_ , coef_ = perceptron(x,y)
```

```python
m = -(coef_[0]/ coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_inp = np.linspace(-3,3,100)
y_inp = m*x_inp + b
```

```python
plt.figure(figsize = (10,6))
plt.plot(x_inp , y_inp , color = 'red' , linewidth = 1)
plt.scatter(x[:,0] , x[:,1] , c=y , cmap='winter' , s =100)
plt.ylim(-3,2)
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
```

```python
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
df = pd.DataFrame(data.data , columns = data.feature_names)
```

```python
data.data
```

```python
df['target'] = data.target
```

```python
x = df.drop(columns= ['target'])
y = df.target
# Split the data
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42, stratify = y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x_train)
xt_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
ior = LogisticRegression()
```

```python
from sklearn.decomposition import PCA
```

```python

pc = PCA(n_components = 2)
X = pc.fit_transform(x_scaled)
xt = pc.transform(xt_scaled)
ior.fit(X,y_train)
```

```python
pd.DataFrame(xt_scaled)
```

```python
ior.coef_[0]
```

```python
m = -1*(ior.coef_[0][0]/ ior.coef_[0][1])
b = -1*(ior.intercept_/ior.coef_[0][1])
```

```python

x_inp1 = np.linspace(x_scaled[:,0].min() - 1,x_scaled[:,0].max() + 1,100)
y_inp1 = m*x_inp1+ b
```
